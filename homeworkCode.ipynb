```python
import matplotlib.pyplot as plt
import odeSolver as os
import numpy as np
```

```python
fprime = lambda t,y : -20*y + 20*np.sin(t) + np.cos(t)
f = lambda t : np.sin(t) + np.exp(-20*t)
```

```python
names = ['euler','rk4','trapezoid']
# hValues = np.array([1/10*(1/2)**x for x in range(6)])
l = len(names)
step = 0.1

ti = 0
tf = 2
yi = 1

values = {}

for i in range(l):
    values[names[i]] = os.solveODE(fprime,ti,tf,yi,names[i],h=step)
```

```python
x = np.linspace(ti,tf)
y = f(x)
```

```python
plt.figure(figsize=(20,10))
design = ['--o','--*','--+']

plt.plot(x,y,label='exact solution')

for i in range(l):
    plt.plot(values[names[i]][0],values[names[i]][1],design[i],label= names[i] +': h = ' + str(step))
    
plt.title("Approximate solution to dy/dt = f(t,y)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
```
