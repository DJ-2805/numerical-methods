```python
import matplotlib.pyplot as plt
import odeSolver as os
import numpy as np
```

***Problem 2***

BVP:
$$ y'' - 2y' + y = x e^x - x $$
Where $x \in [0,2], y(0)=0, y(2) = -4, h = 0.2 $

The exact solution is
$$ y(x) = \frac{1}{6}x^3e^x - \frac{5}{3}xe^x + 2xe^x - x - 2$$

```python
p = -2
q = 1
fprime = lambda x : x*np.exp(x)-x
f = lambda x : 1/6*x**3*np.exp(x) - 5/3*x*np.exp(x) + 2*np.exp(x) - x - 2
```

```python
ta = 0
ya = 0

tb = 2
yb = -4
h = 0.2

y,x = os.solveODE2(fprime,p,q,ta,ya,tb,yb,h=h)
```

```python
plt.figure(figsize=(10,10))

plt.plot(x,f(x),label='exact Solution')
plt.plot(x,y,'--*',label= 'FDM Solution')
    
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend()
```
