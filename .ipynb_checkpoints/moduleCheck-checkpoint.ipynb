```python
import matplotlib.pyplot as plt
import numericalMethods as nm
```

```python
def euler(f,ti,tf,yi,n):
    h = (tf-ti)/n
    y = np.zeros(n+1)
    t = np.zeros(n+1)
    
    y[0] = yi
    t[0] = ti
    
    for i in range(n):
        t[i+1] = t[i] + h
        y[i+1] = y[i] + h*f(t[i],y[i])
        
    return t,y
```

```python
f = lambda t,y : -5*y + 5*t**2 + 2*t
```

```python
nValues = np.array([10,20,40])
ti = 0
tf = 1
yi = 1/3
l = len(nValues)

tValues = {}
yValues = {}

for i in range(l):
    tValues[i], yValues[i] = euler(f,ti,tf,yi,nValues[i])
```

```python
for i in range(l):
    plt.plot(tValues[i],yValues[i])
    
plt.title("Approximate solution to dy/dt = -y obtained with Euler's method")
plt.xlabel('t')
plt.ylabel('y(t)')
```
