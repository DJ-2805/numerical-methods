```python
import matplotlib.pyplot as plt
import odeSolver as os
import numpy as np
```

```python
f = lambda t,y : 4*y*(1-y)
```

```python
hValues = np.array([1/10*(1/2)**x for x in range(6)])
ti = 0
tf = 1
yi = 0.1
l = len(hValues)

tEuler = {}
yEuler = {}

tHeun = {}
yHeun = {}

for i in range(l):
    tEuler[i], yEuler[i] = os.euler(f,ti,tf,yi,h=hValues[i])
    tHeun[i], yHeun[i] = os.heun(f,ti,tf,yi,h=hValues[i])
```

```python
plt.figure(figsize=(20,10))
for i in range(l):
    plt.plot(tEuler[i],yEuler[i],'--+',label='Euler: h = ' + str(hValues[i]))
    plt.plot(tHeun[i],yHeun[i],'--', label='Heun: h = ' + str(hValues[i]))
    
plt.title("Approximate solution to dy/dt = f(t,y)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.savefig('approxPlot.png')
```

```python
file = open('data.txt','a')

info = 'Date: 2019-10-17'
info += '\nValues collected by David James\n'
file.write(info)

for i in range(l):
    info = '\nh = ' + str(hValues[i])
    info += '\nEuler:t     y          Heun:t     y'
    np.savetxt(file,
               list(zip(tEuler[i],yEuler[i],tHeun[i],yHeun[i])),
               header=info,
               fmt='%10.5f')
```
