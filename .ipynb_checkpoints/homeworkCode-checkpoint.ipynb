## Problem 1

Consider the IVP:
$$ y' = 5y + 5y^2 + 2t, 0 \leq t \leq 1, y(0) = 1/3, h = 0.1 $$
The exact solution is
$$ y(t) = t^2 + \frac{1}{3}e^{-5t} $$

```python
import matplotlib.pyplot as plt
import odeSolver as os
import numpy as np
```

```python
fprime = lambda t,y : -5*y + 5*t**2 + 2*t
f = lambda t : t**2 + 1/3*np.exp(-5*t)
```

```python
names = ['ab2','am2']
l = len(names)
step = 0.1

ti = 0
tf = 1
yi = 1/3

values = {}

for i in range(l):
    values[names[i]] = os.solveODE(fprime,ti,tf,yi,names[i],h=step)
```

```python
x = np.linspace(ti,tf)
y = f(x)
```

***Part 1***

```python
plt.figure(figsize=(20,10))
design = ['--o','--*','--+','--o','--*']

plt.plot(x,y,label='exact solution')
plt.plot(values['ab2'][0],values['ab2'][1],design[0],label= names[0] +': h = ' + str(step))
    
plt.title("Approximate solution to dy/dt = f(t,y)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
```

***Part 3***

```python
plt.figure(figsize=(20,10))
design = ['--o','--*','--+','--o','--*']

plt.plot(x,y,label='exact solution')
plt.plot(values['am2'][0],values['am2'][1],design[0],label= names[0] +': h = ' + str(step))
    
plt.title("Approximate solution to dy/dt = f(t,y)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
```

## Problem 2

The higher order DE is
$$ y''' + 2y'' - y' -2y = e^t, 0 \leq t \leq 3, y(0) = 1, y'(0) = 2, y''(0) = 0, h = 0.1 $$

The exact solution is
$$ y(t) = \frac{43}{36}e^t + \frac{1}{4}e^{-t} - \frac{4}{9}e^{-2t} + \frac{1}{6}te^t $$

***Part 1***

Let $v_1 = y$, $v_2 = y'$, and $v_3 = y''$, therefore, applying this would lead to

\begin{align}
    v_1'  &= y'   &= v_2 \\
    v_2'  &= y''  &= v_3 \\
    v_3'  &= y''' &= e^t - 2v_3 + v_2 + 2v_1
\end{align}

Hence the matrix form of this would be
\begin{equation*}
\frac{d\vec{v}}{dt} =
\left(
\begin{matrix}
    0 & 1 & 0 \\
    0 & 0 & 1 \\
    2 & 1 & -2
\end{matrix}
\right) \vec{v} + 
\left(
\begin{matrix}
    0 \\ 0 \\ e^t
\end{matrix}
\right)
\end{equation*}

***Part 2***

RK-2 in matrix form is as follows 


\begin{align}
\vec{s_1} &= h*f(t,\vec{y}) //
\vec{s_2} &= h*f(t + \frac{h}{2},\vec{y} + \frac{s_1}{2})
\vec{y_{i+1} &= \vec{y_i} + s_2
\end{align}

***Part 3***

```python
def gprime(t,y):
    a = np.zeros((3,3))
    a[0,1] = 1
    a[1,2] = 1
    a[2,0] = 2
    a[2,1] = 1
    a[2,2] = -2
    
    b = np.zeros(3)
    b[2] = np.exp(t)
    return a.dot(y) + b
    
    
g = lambda t : 43/36*np.exp(t) + 1/4*np.exp(-t) - 4/9*np.exp(-2*t) + 1/6*t*np.exp(t)
```

```python
ti = 0
tf = 3
yi = np.array([1,2,0])
step = 0.1
name = 'heun'

u,t = os.solveSYS(gprime,ti,tf,yi,m=3,h=step)
```

```python
x = np.linspace(ti,tf)
y = g(x)
```

```python
plt.figure(figsize=(20,10))

plt.plot(x,y,label='exact solution')
plt.plot(t,u[0],'--o',label= name + ': h = ' + str(step))
    
plt.title("Approximate solution to dy/dt = f(t,y)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
```
